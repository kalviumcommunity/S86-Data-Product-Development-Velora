# Time Series Analysis

This notebook computes weekly and monthly resampling, rolling averages, month-over-month growth, cumulative revenue, and a short business interpretation.

The repository does not include a long daily revenue series, so the notebook uses the same reproducible daily generator as the script to demonstrate the required time-series methods.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent

if str(ROOT / 'scripts') not in sys.path:
    sys.path.append(str(ROOT / 'scripts'))

from rolling_metrics import (
    build_daily_data,
    compute_cumulative_revenue,
    compute_month_over_month,
    compute_rolling_averages,
    resample_time_periods,
    save_plots,
    summarize_trend,
)

OUTPUT_DIR = ROOT / 'output'
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

df = build_daily_data()
df.head()
```

```python
weekly, monthly, comparison = resample_time_periods(df)
df = compute_rolling_averages(df)
df = compute_cumulative_revenue(df)
mom_change = compute_month_over_month(df)

print('Weekly resample:')
print(weekly.head().to_string())
print('\nMonthly resample:')
print(monthly.head().to_string())
print('\nComparison summary:')
print(comparison.to_string())
print('\nHighest average bucket revenue:', comparison['avg_bucket_revenue'].idxmax())
```

```python
save_plots(df)

trend_analysis = summarize_trend(df, mom_change)
(OUTPUT_DIR / 'trend_analysis.txt').write_text(trend_analysis, encoding='utf-8')
print(trend_analysis)
print(f'Total revenue: ${df["cumulative_revenue"].iloc[-1]:,.0f}')
```
